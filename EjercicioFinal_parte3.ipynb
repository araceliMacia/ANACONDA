## Parte 3

Podemos descargar datos de empresas del dataset Forbes2000 (paquete HSAUR) mediante get_rdataset(). Encontraremos un ranking de empresas categorizadas por sector, y con datos de su país, ventas, beneficios, recursos y valor de mercado.



```python
import statsmodels.api as sm

import pandas as pd
import numpy as np

forbes = sm.datasets.get_rdataset("Forbes2000","HSAUR")
print forbes.__doc__
Datosforbes = forbes.data
Datosforbes.head(2)
```

*  **Seleccionar del dataframe los nombres de empresa que incluyen "elect" (sin tener en cuenta mayúsculas). Intentamos con esto obtener empresas “del sector de la electricidad” de forma aproximada**

```python
electricas = Datosforbes [Datosforbes.name.str.contains("elect", case=False)==True]
electricas = pd.DataFrame(electricas)
print electricas.name
```

* **Utilizando como índice el país y la categoría y ordenándolo (sort_index), encontrar el valor de mercado de las empresas japonesas y estadounidenses.**

```python
byCountry = electricas.set_index(["country", "category"])
byCountry.sort_index(inplace=True)
byCountry
```

```python

MercadoJapUSA= byCountry.loc[['Japan', 'United States'], ["name","marketvalue"]]
MercadoJapUSA.tail(20)
```

* **Sobre el mismo subconjunto de empresas eléctricas, obtener el país donde el sector de utilities tiene mayor valor de mercado medio en sus empresas.**

```python

print "media por pais ", electricas[electricas["category"]=="Utilities"].groupby(['country'])['marketvalue'].mean()
print "-----"
print "valor maximo Es : ", electricas[electricas["category"]=="Utilities"].groupby(['country'])['marketvalue'].mean().max()
print "-----"
print  "pais con maximo valor:" , electricas[electricas["category"]=="Utilities"].groupby(['country'])['marketvalue'].mean().argmax()

#byCountry.loc[(slice(None), ['Utilities']),:]['marketvalue']
#byCountry.iloc[byCountry.index.isin(['Utilities'], level=1)]
```
